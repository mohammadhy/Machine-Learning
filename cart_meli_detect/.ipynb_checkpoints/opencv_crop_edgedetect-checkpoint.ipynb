```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
def increase_brightness(img, value=30):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)

    lim = 255 - value
    v[v > lim] = 255
    v[v <= lim] += value

    final_hsv = cv2.merge((h, s, v))
    img = cv2.cvtColor(final_hsv, cv2.COLOR_HSV2BGR)
    return img
```

```python
def procces_image(path_image):
    
    image = cv2.imread(path_image)
    
    print(image.shape)
    
    
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    print(gray.shape)
    edged = cv2.Canny(gray, 30, 200)

    plt.imshow(edged)
    
    contours, hierarchy = cv2.findContours(edged, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    cnts = contours[0]
    
    x,y,w,h = cv2.boundingRect(cnts)
    
    print(x, y, w, h)
    
    cv2.circle(image,(x,y), 20, (0,0,0), -1)

    cv2.circle(image,(x+w,y), 20, (0,0,0), -1)

    cv2.circle(image,(x+w,y+h), 20, (0,0,0), -1)

    cv2.circle(image,(x,y+h), 20, (0,0,0), -1)
    
    plt.imshow(image)
    cv2.drawContours(image, contours, -1, (255, 255, 255), 3)
    
    image = image[y:y+h, x:x+w]
    
    plt.imshow(image)  
    return image
```

```python
def extract_to_read_by_pytesseract(image):
    num_meli = image[130:180, 500:]
    name_meli = image[180:230, 500:]
    lastname_meli = image[230:280, 500:]
    birthday_meli = image[280:320, 400:]
    fathername_meli = image[32:380, 500:]
    expire_meli = image[380:420, 400:]
    return num_meli, name_meli, lastname_meli, birthday_meli, fathername_meli, expire_meli
```
