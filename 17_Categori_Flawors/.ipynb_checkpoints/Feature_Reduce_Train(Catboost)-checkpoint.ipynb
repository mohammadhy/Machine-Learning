```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import numpy as np
import pandas as pd
import cv2
import os
import h5py
import mahotas
import pickle
```

```python
labels = os.listdir('data/')
len(labels)
```

```python
x = [] ## For X
y = [] ## For Label
for label in labels:
                ## choose All Label And Their Photos Inside Lables
    dir = os.path.join('data', label)
    #print(dir)
    current_label = label
    #print('Current Folder: ' + current_label)
    for k in range(1, 81):
                ## Now For Read All Photo In Directory We Need To Go On Specific Dir And 
                ## MaKe Loop On All Photo 80 Photo In Each Dir
        #print('Current Image: ' + str(x))
        file = os.path.join(dir, str(k) + '.jpg')
                ## Look At the File And See Loop On All Label And All Photo In Dir Each Label
        #print(file)
                ## Now Read The Each Photo With CV2
        image = cv2.imread(file)
        image = cv2.resize(image, (500, 500))
                ## Resize Theme Into One Scale 500 * 500 & Chage Channel
                ## Convert from BGR to HSV color-space
        image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
                ##Histogram
        hist = cv2.calcHist([image], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
        cv2.normalize(hist, hist)
        histogram_features = hist.flatten()
                ## As Same As Histogram But Use Haralick
        image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        hue_features = cv2.HuMoments(cv2.moments(image_gray)).flatten()
        haralick_features = mahotas.features.haralick(image_gray).mean(axis=0)
                ## Horizantally Numpy 3 Feature Append Into Each Other
        current_features = np.hstack([histogram_features, hue_features, haralick_features])
        
        x.append(current_features)
        y.append(current_label)
```

```python
np.shape(x)
```

```python
np.shape(y)
```

```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder, StandardScaler
min_max = MinMaxScaler()
x_minmax = min_max.fit_transform(x)
## You Can Use Both Or One Of Theme: 
stn = StandardScaler()
x_minmax_s = stn.fit_transform(x_minmax)
```

```python
lb = LabelEncoder()
y_label = lb.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_minmax_s, y_label, random_state=125)
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import classification_report
```

```python
ct = CatBoostClassifier(iterations=400 ,depth=5 ,random_state=125)
```

```python
ct.fit(x_train, y_train)
```

```python
y_p = ct.predict(x_test)
print(classification_report(y_p, y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
rn = RandomForestClassifier(random_state=125, n_estimators=400)
rn.fit(x_train, y_train)
```

```python
y_r = rn.predict(x_test)
```

```python
print(classification_report(y_r, y_test))
```
