```python
from nltk.tokenize import word_tokenize
from spellchecker import SpellChecker
```

```python
word_tokenize(text)
```

```python
check = SpellChecker()
```

```python
def word_recomend(sentence):
    correct_word = []
    not_exist = 'Nan'
    for word in word_tokenize(sentence):
        recomend = list(check.candidates(word))
        print(f'The Real Word {[word]} And Recomended Word :{recomend}')
        index = input('Enter The Correct word From Recommended_list: ')
        #print(index)
        if index != '':
            chosen_word = (recomend[int(index)])
            print(type(chosen_word))
            correct_word.append(chosen_word)
        else:
            correct_word.append(word)
    return correct_word
```

```python
word_recomendation = word_recomend(text)
```

```python
old_word = word_tokenize(text)
```

```python
def return_true(old, new):
    final = old +  [x for x in new if x not in new]
    return final
```

```python
new_text = return_true(word_recomendation, old_word)
```
