```python
import os
import glob
import shutil
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
```

```python
class_name = ['0', '1', '2', '3', '4', '5']
l = 0
for k in range(0, 6):
    os.makedirs('data/' + class_name[l])
    current_path = 'data/' + class_name[l]
    for img_patg in glob.glob('datasets/*_{}.png'.format(l)):
        print(img_patg)
        shutil.move(img_patg, current_path)
    l +=1
```

```python
img = cv2.imread('data/2/0_2.png')
```

```python
img.shape
```

```python
num = []
labels = os.listdir('data/')
for label in labels:
    print(label)
    path = os.path.join('data/', str(label) + '/')
    print(type(path))
    class_ = labels.index(label)
    for image in os.listdir(path):
        try:
            img = cv2.imread(os.path.join(path, image))
            img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img_resize = cv2.resize(img_gray, (100, 100))
            num.append([img_resize, class_])
        except Exception as e:
            print(e)
```

```python
num = np.array(num)
```

```python
plt.imshow(num[8000][0])
```

```python
x = []
y = []
for feature, label in num:
    x.append(feature)
    y.append(label)
```

```python
x = np.array(x)
y = np.array(y)
```

```python
x.shape
```

```python
x = x.reshape(-1, 1)
stn = StandardScaler()
x = stn.fit_transform(x)
```

```python
x.shape
```

```python
x = x.reshape(12005, 100, 100, 3)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x ,y, test_size = 0.2 ,random_state = 123)
```

```python
#x_train = np.expand_dims(x_train, axis=3)
#x_test = np.expand_dims(x_test, axis=3)
```

```python
x_test.shape
```

```python
x_train.shape
```

## ConvModel:

```python
model = Sequential()
model.add(layers.Conv2D(32, 3,padding = 'same' ,input_shape=(100, 100, 3)))
model.add(layers.MaxPooling2D(2))
model.add(layers.MaxPooling2D(2))

model.add(layers.Dropout(0.5))

model.add(layers.Conv2D(64, 3,padding = 'same'))
model.add(layers.MaxPooling2D(2))
model.add(layers.MaxPooling2D(2))

model.add(layers.Dropout(0.5))

model.add(layers.Conv2D(64, 3,padding = 'same'))
model.add(layers.MaxPooling2D(2))
model.add(layers.MaxPooling2D(2))

model.add(layers.Dropout(0.5))

model.add(layers.Flatten())

model.add(layers.Dropout(0.5))

model.add(layers.Dense(64, activation = 'elu'))

model.add(layers.Dense(32, activation = 'elu'))

model.add(layers.Dense(16, activation = 'elu'))

model.add(layers.Dense(6, activation = 'softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', metrics=['accuracy'], loss = 'sparse_categorical_crossentropy' )
```

```python
res = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test), batch_size = 100 )
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```
