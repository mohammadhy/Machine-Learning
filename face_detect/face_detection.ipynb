```python
names = ['Unkhonwn', 'adrian', 'adrian', 'matew']
encods = []
encods.append()
encods.append(b2_code)
encods.append(b1_code)
encods.append(m_code)
```

```python
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()

    faces_loc = fr.face_locations(frame)
    faces_encode = fr.face_encodings(frame, faces_loc)
    
    for encodeFace, faceLoc in zip(faces_encode, faces_loc):
        matches = fr.compare_faces(encods, encodeFace)
        print(matches)
        
        face_distances = fr.face_distance(encods, encodeFace)
        matchIndex = np.argmin(face_distances)
        
        print(matchIndex)
        if matches[matchIndex]:
            name = names[matchIndex]
            print(name)
            y1,x2,y2,x1 = faceLoc
            
            cv2.rectangle(frame,(x1,y1), (x2,y2), (255,0,255), 2)
            # for write Text
            cv2.rectangle(frame, (x1,y2-35), (x2,y2), (255,0,255), cv2.FILLED)
            cv2.putText(frame,name,
                        (x1+6, y2-6), cv2.FONT_HERSHEY_COMPLEX, 0.5, (255, 255, 255), 1)

    cv2.imshow('webcam', frame)
    
    if cv2.waitKey(1) == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
